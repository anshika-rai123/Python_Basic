5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each


```python
# Create a list and a tuple
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print("Original List:", my_list)
print("Original Tuple:", my_tuple)

# Modify list element
my_list[0] = 100
print("Modified List:", my_list)

# Try modifying tuple element
try:
    my_tuple[0] = 100   # This will raise an error
except TypeError as e:
    print("Error when modifying tuple:", e)
```

6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
# Function to demonstrate mutable and immutable behavior
def modify(mutable_list, immutable_num):
    # Modify list (mutable)
    mutable_list.append(100)

    # Modify integer (immutable)
    immutable_num += 10

    print("Inside function:")
    print("List:", mutable_list)
    print("Number:", immutable_num)

# Outside function
my_list = [1, 2, 3]
my_num = 50

print("Before function call:")
print("List:", my_list)
print("Number:", my_num)

modify(my_list, my_num)

print("\nAfter function call:")
print("List:", my_list)  # Changed because list is mutable
print("Number:", my_num) # Unchanged because int is immutable
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
even_sum = sum(i for i in range(1, 51) if i % 2 == 0)
print("Sum of even numbers:", even_sum)
```

14. Write a program to reverse a string using a while loop.


```python
s = input("Enter a string: ")
i = len(s) - 1
rev = ""
while i >= 0:
    rev += s[i]
    i -= 1
print("Reversed string:", rev)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
n = int(input("Enter a number: "))
fact = 1
while n > 0:
    fact *= n
    n -= 1
print("Factorial:", fact)
```
